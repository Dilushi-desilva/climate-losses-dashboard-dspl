```python
df_country_year = pd.read_excel("C:/Users/dilus/Downloads/eea_t_losses-from-climate-related-extremes_p_1980-2024_v01_r01/eea_t_losses-from-climate-related-extremes_p_1980-2024_v01_r01/Data_EconomicLosses 2024_rev2025Q4_rev1.xlsx", 
                                 sheet_name='Country_year', 
                                 header=1)

print(df_country_year.shape)
df_country_year.head()
```

```python
year_cols = [c for c in df_country_year.columns if isinstance(c, (int, float)) and 1980 <= c <= 2024]
```

```python
df_country_year = df_country_year[
    (df_country_year['price level'] == 2024) &
    (df_country_year['Hazard Type'] == 'Weather- and climate related extremes') &
    (df_country_year['Category'] == 'Total losses')
][['ISO2', 'Name'] + year_cols].copy()
```

```python
df_country_year = df_country_year.reset_index(drop=True)

print(df_country_year.shape)
df_country_year.head()
```
